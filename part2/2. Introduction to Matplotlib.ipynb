### Multi-line plots

```python
x = np.arange(10)
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.plot(x, x**4)
plt.show()
```

### Grid, axes, labels

```python
x = np.arange(10)
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.plot(x, x**4)
plt.grid(True)
plt.savefig("test.png")
plt.show()
```

The statement _plt.savefig('test.png')_ saves the image in the current directory.




```python
x = np.arange(10)
plt.plot(x, x**2, label="x**2")
plt.plot(x, x**3, label="x**3")
plt.plot(x, x**4, label="x**4")
plt.grid(True)
# plt.legend()
# plt.xlim(0,5)
# plt.ylim(0,2000)
plt.title("Simple plot")
plt.ylabel("y=f(x)")
plt.savefig("test.png")
plt.show()
```

### Colours, styles, markers



```python
x = np.arange(10)
plt.plot(x, x**2, color = "g", linestyle = "--", label="x**2" )
plt.plot(x, x**3, color = "y", label="x**3")
plt.plot(x, x**4, color = "r", label="x**4")
plt.grid(True)
# plt.legend()
# plt.xlim(0,5)
# plt.ylim(0,2000)
plt.title("Simple plot")
plt.ylabel("y=f(x)")
plt.savefig("test.png")
plt.show()
```

## Object-oriented plotting (another way)

```python
fig = plt.figure()
ax = fig.add_subplot(111) ## one row and one column
ax.plot([1, 2, 3, 4], 
        [10, 20, 25, 30], 
        color='lightblue', 
        linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], 
           [11, 25, 9, 26], 
           c=[1, 2, 3, 5], 
           marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
```

### This is the same as: 



```python
plt.plot([1, 2, 3, 4], 
         [10, 20, 25, 30], 
         color='lightblue', 
         linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], 
            [11, 25, 9, 26], 
            c=[1, 2, 3, 5], 
            marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

While very simple plots, with short scripts would benefit from the conciseness of the pyplot implicit approach, when doing more complicated plots, or working within larger scripts, you will want to explicitly pass around the Axes and/or Figure object to operate upon.

The advantage of keeping which axes we're working with very clear in our code will become more obvious when we start to have multiple axes in one figure.

## Scatterplot

```python
# help(plt.scatter)
```

```python
# Create a scatterplot
plt.scatter(x = df.age, 
            y = df.VO2max, 
            s = 20, 
            c = df.gender)

# Set the title and labels
plt.xlabel('Age')
plt.ylabel('VO2 max')
plt.title('VO2 max per age')
# plt.ylim(0)
plt.show()
```

```python
# this is the same as
# notice using data argument
fig, ax = plt.subplots()

# Plot the "centerline" with `plot`
ax.scatter('age', 'VO2max', 
           s = 20,
           c = 'gender',
           data=df)
ax.set_xlabel("age")
ax.set_ylabel("VO2max")
plt.savefig("test.png")

plt.show()
```
